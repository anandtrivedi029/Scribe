```python

# SCRIBE: Complete Training

#!pip install -q sentence-transformers

import torch, torch.nn as nn, torch.nn.functional as F
import pandas as pd, numpy as np, math, time, os, gc
from collections import defaultdict
from sentence_transformers import SentenceTransformer
from tqdm.auto import tqdm

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Device: {device} ({torch.cuda.get_device_name(0) if device.type=='cuda' else 'cpu'})")


# CONFIG

TRAIN_LIMIT = 60000
TEST_LIMIT = 10000
MAX_SENTS = 15
BATCH_SIZE = 256
NUM_EPOCHS = 40
LR = 5e-4  #1e-3
HIDDEN_DIM = 256
NUM_SLOTS = 16
NUM_HEADS = 4
READ_HOPS = 2
CONSOL_EVERY = 4
DROPOUT = 0.1
USE_AMP = True

os.makedirs("experiments", exist_ok=True)


# LOAD & ENCODE DATASET

print("\n[1/3] Loading parquet...")
df = pd.read_parquet("unified_reasoning_dataset.parquet")

#train_df = df[df['split']=='train'].head(TRAIN_LIMIT).reset_index(drop=True)
#test_df = df[df['split']=='test'].head(TEST_LIMIT).reset_index(drop=True)
#train_df = df[df['split']=='train'].sample(n=TRAIN_LIMIT, random_state=42).reset_index(drop=True)
#test_df = df[df['split']=='test'].sample(n=TEST_LIMIT, random_state=42).reset_index(drop=True)

train_pw = df[(df['split']=='train') & (df['task']=='proofwriter')].sample(n=20000, random_state=42)
train_ba = df[(df['split']=='train') & (df['task']=='babi')]  # Take all bAbI available
train_df = pd.concat([train_pw, train_ba.sample(n=min(20000, len(train_ba)), random_state=42)]).sample(frac=1, random_state=42).reset_index(drop=True)

test_pw = df[(df['split']=='test') & (df['task']=='proofwriter')].sample(n=5000, random_state=42)
test_ba = df[(df['split']=='test') & (df['task']=='babi')]
test_df = pd.concat([test_pw, test_ba.sample(n=min(5000, len(test_ba)), random_state=42)]).sample(frac=1, random_state=42).reset_index(drop=True)


vocab = {a:i for i,a in enumerate(df['answer'].unique())}
del df; gc.collect()

print(f"Train: {len(train_df)} | Test: {len(test_df)} | Classes: {len(vocab)}")
print(f"Tasks: {train_df['task'].value_counts().to_dict()}")

print("\n[2/3] Encoding sentences (one-time)...")
encoder = SentenceTransformer('all-MiniLM-L6-v2').to(device)

def encode_dataset(dataframe, encoder, max_sents=15, batch_size=512):
    N = len(dataframe)
    questions = dataframe['question'].tolist()
    contexts = dataframe['context'].tolist()

    # Collect all sentences flat
    all_sents = []
    sents_per_sample = []
    for ctx in contexts:
        s = [x.strip() for x in ctx.split('.') if len(x.strip()) > 3]
        if not s: s = [ctx[:100]]
        s = s[:max_sents]
        sents_per_sample.append(len(s))
        all_sents.extend(s)

    # Encode in bulk
    with torch.no_grad():
        q_vecs = encoder.encode(questions, convert_to_tensor=True, batch_size=batch_size, show_progress_bar=True).clone().cpu()
        s_vecs_flat = encoder.encode(all_sents, convert_to_tensor=True, batch_size=batch_size, show_progress_bar=True).clone().cpu()

    # Assemble
    s_vecs = torch.zeros(N, max_sents, 384)
    s_masks = torch.zeros(N, max_sents)
    idx = 0
    for i in range(N):
        n = sents_per_sample[i]
        s_vecs[i,:n] = s_vecs_flat[idx:idx+n]
        s_masks[i,:n] = 1.0
        idx += n

    del s_vecs_flat; gc.collect()
    return q_vecs, s_vecs, s_masks, sents_per_sample

train_q, train_s, train_m, train_nsents = encode_dataset(train_df, encoder, MAX_SENTS)
test_q, test_s, test_m, test_nsents = encode_dataset(test_df, encoder, MAX_SENTS)

# Free encoder
del encoder; gc.collect(); torch.cuda.empty_cache()
print("Encoding done, encoder freed.")

# Prepare labels
train_answers = [vocab.get(a,0) for a in train_df['answer']]
test_answers = [vocab.get(a,0) for a in test_df['answer']]
train_tasks = train_df['task'].tolist()
test_tasks = test_df['task'].tolist()

# ============================================================================
# MODEL
# ============================================================================
class MemoryReader(nn.Module):
    def __init__(self, D, heads=4, hops=2, drop=0.1):
        super().__init__()
        self.hops = hops
        self.attn = nn.ModuleList([nn.MultiheadAttention(D, heads, batch_first=True, dropout=drop) for _ in range(hops)])
        self.norm = nn.ModuleList([nn.LayerNorm(D) for _ in range(hops)])
        self.gate = nn.ModuleList([nn.Linear(D*2, D) for _ in range(hops)])
    def forward(self, query, mem, mask):
        ig = (mask==0); state = query
        for i in range(self.hops):
            q = self.norm[i](state).unsqueeze(1)
            o, _ = self.attn[i](q, mem, mem, key_padding_mask=ig)
            o = o.squeeze(1)
            g = torch.sigmoid(self.gate[i](torch.cat([state,o],-1)))
            state = g*o + (1-g)*state
        return state

class SCRIBE(nn.Module):
    def __init__(self, in_d=384, D=192, slots=16, heads=4, rhops=2, consol=4, n_cls=100, drop=0.1):
        super().__init__()
        self.D=D; self.S=slots; self.consol=consol
        self.s_enc = nn.Sequential(nn.Linear(in_d,D),nn.GELU(),nn.Dropout(drop),nn.Linear(D,D),nn.LayerNorm(D))
        self.q_enc = nn.Sequential(nn.Linear(in_d,D),nn.GELU(),nn.Dropout(drop),nn.Linear(D,D),nn.LayerNorm(D))
        self.sentinel = nn.Parameter(torch.randn(D)*0.02)
        self.mem_attn = nn.MultiheadAttention(D, heads, batch_first=True, dropout=drop)
        self.mem_norm = nn.LayerNorm(D)
        self.w_gate = nn.Sequential(nn.Linear(D*2,D),nn.GELU(),nn.Linear(D,D),nn.Sigmoid())
        self.conf = nn.Sequential(nn.Linear(D*2,1),nn.Sigmoid())
        self.rev_proj = nn.Linear(D,D)
        self.rev_gate = nn.Sequential(nn.Linear(D*2,D),nn.Sigmoid())
        self.consolidation = nn.TransformerEncoderLayer(D, heads, D*2, drop, batch_first=True)
        self.reader = MemoryReader(D, heads, rhops, drop)
        self.head = nn.Sequential(nn.LayerNorm(D),nn.Linear(D,D),nn.GELU(),nn.Dropout(drop),nn.Linear(D,n_cls))
        self._init()
    def _init(self):
        for m in self.modules():
            if isinstance(m, nn.Linear):
                nn.init.xavier_uniform_(m.weight)
                if m.bias is not None: nn.init.zeros_(m.bias)
    def forward(self, sents, query, smask=None):
        B,N,_ = sents.shape; dev=sents.device; S=self.S; D=self.D
        if smask is None: smask=torch.ones(B,N,device=dev)
        mx = min(N,S-1)
        enc = self.s_enc(sents[:,:mx])
        mem = torch.zeros(B,S,D,device=dev)
        mm = torch.zeros(B,S,device=dev)
        mem[:,0,:] = self.sentinel.unsqueeze(0).expand(B,-1)
        mm[:,0] = 1.0
        wp = 1
        for t in range(mx):
            act = smask[:,t]
            if act.sum()==0: continue
            if wp>=S: break
            st = enc[:,t,:]
            q = self.mem_norm(st).unsqueeze(1)
            ctx,_ = self.mem_attn(q, mem, mem, key_padding_mask=(mm==0))
            ctx = ctx.squeeze(1)
            comb = torch.cat([st,ctx],-1)
            wg = self.w_gate(comb)
            wv = wg*st + (1-wg)*ctx
            cf = self.conf(comb)
            # Revision
            rs = self.rev_proj(st)
            sim = torch.bmm(mem, rs.unsqueeze(-1)).squeeze(-1)/math.sqrt(D)
            rw = torch.sigmoid(sim)*mm
            re = rs.unsqueeze(1).expand_as(mem)
            ri = torch.cat([mem,re],-1)
            rg = self.rev_gate(ri)
            rst = (cf*act.unsqueeze(-1)).unsqueeze(1)
            rsc = rw.unsqueeze(-1)*rst*0.3
            mem = mem + rg*(re-mem)*rsc
            # Write
            ws = cf.squeeze(-1)*act
            sc = wv*ws.unsqueeze(-1)
            wm = torch.zeros(B,S,1,device=dev); wm[:,wp,:]=1.0
            mem = mem*(1-wm) + sc.unsqueeze(1)*wm
            mu = torch.zeros(B,S,device=dev); mu[:,wp]=ws
            mm = mm + mu
            wp += 1
            if (t+1)%self.consol==0 and t>0:
                mem = self.consolidation(mem, src_key_padding_mask=(mm<0.1))
        mem = self.consolidation(mem, src_key_padding_mask=(mm<0.1))
        qr = self.q_enc(query)
        rs = self.reader(qr, mem, mm)
        return self.head(rs)


# INIT

model = SCRIBE(384, HIDDEN_DIM, NUM_SLOTS, NUM_HEADS, READ_HOPS, CONSOL_EVERY, len(vocab), DROPOUT).to(device)
params = sum(p.numel() for p in model.parameters() if p.requires_grad)
print(f"\n[3/3] SCRIBE: {params:,} params")


# TRAIN

optimizer = torch.optim.AdamW(model.parameters(), lr=LR, weight_decay=0.01)
N = len(train_df)
steps = NUM_EPOCHS * ((N+BATCH_SIZE-1)//BATCH_SIZE)
scheduler = torch.optim.lr_scheduler.OneCycleLR(optimizer, max_lr=LR, total_steps=steps, pct_start=0.1)

print(f"\nTraining: {N} samples, batch={BATCH_SIZE}, epochs={NUM_EPOCHS}")
print("="*80)

history = []
best_acc = 0

for epoch in range(NUM_EPOCHS):
    t0 = time.time()
    model.train()
    perm = torch.randperm(N)
    nb = (N+BATCH_SIZE-1)//BATCH_SIZE
    scaler = torch.amp.GradScaler('cuda', enabled=USE_AMP)
    tloss=0; correct=0; total=0

    for bi in tqdm(range(nb), desc=f"Epoch {epoch+1}", leave=False):
        s,e = bi*BATCH_SIZE, min((bi+1)*BATCH_SIZE, N)
        ix = perm[s:e]
        q = train_q[ix].to(device)
        sv = train_s[ix].to(device)
        mk = train_m[ix].to(device)
        ans = torch.tensor([train_answers[i] for i in ix], dtype=torch.long, device=device)

        optimizer.zero_grad()
        with torch.amp.autocast('cuda', enabled=USE_AMP):
            logits = model(sv, q, mk)
            loss = F.cross_entropy(logits, ans)
        scaler.scale(loss).backward()
        scaler.unscale_(optimizer)
        torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
        scaler.step(optimizer); scaler.update(); scheduler.step()

        tloss += loss.item()
        correct += (logits.argmax(1)==ans).sum().item()
        total += len(ans)
        del q,sv,mk,ans,logits,loss

    # Eval
    model.eval()
    tc=0; tt=0
    per_task = defaultdict(lambda:{'c':0,'t':0})
    per_depth = defaultdict(lambda:{'c':0,'t':0})

    with torch.no_grad():
        Nt = len(test_df)
        for bi in range((Nt+BATCH_SIZE-1)//BATCH_SIZE):
            s,e = bi*BATCH_SIZE, min((bi+1)*BATCH_SIZE, Nt)
            q = test_q[s:e].to(device)
            sv = test_s[s:e].to(device)
            mk = test_m[s:e].to(device)
            ans = torch.tensor(test_answers[s:e], dtype=torch.long, device=device)
            with torch.amp.autocast('cuda', enabled=USE_AMP):
                logits = model(sv, q, mk)
            match = (logits.argmax(1)==ans)
            tc += match.sum().item(); tt += len(ans)
            for i in range(len(match)):
                ri = s+i
                tk = test_tasks[ri]
                per_task[tk]['c'] += match[i].item(); per_task[tk]['t'] += 1
                ns = test_nsents[ri]
                d = 'easy(1-3)' if ns<=3 else 'med(4-6)' if ns<=6 else 'hard(7+)'
                per_depth[d]['c'] += match[i].item(); per_depth[d]['t'] += 1
            del q,sv,mk,ans,logits

    tr_acc = correct/total*100
    te_acc = tc/tt*100
    elapsed = time.time()-t0

    task_acc = {t:d['c']/d['t']*100 for t,d in per_task.items()}
    depth_acc = {d:v['c']/v['t']*100 for d,v in per_depth.items() if v['t']>0}

    print(f"\nEpoch {epoch+1:2d}/{NUM_EPOCHS} ({elapsed:.0f}s) | Loss: {tloss/nb:.4f} | Train: {tr_acc:.1f}% | Test: {te_acc:.1f}%")
    tk_str = " | ".join(f"{t}:{a:.1f}%" for t,a in sorted(task_acc.items()))
    dp_str = " | ".join(f"{d}:{a:.1f}%" for d,a in sorted(depth_acc.items()))
    print(f"  Tasks: {tk_str}")
    print(f"  Depth: {dp_str}")

    if te_acc > best_acc:
        best_acc = te_acc
        torch.save({'epoch':epoch+1,'state':model.state_dict(),'test_acc':te_acc,
                     'task_acc':task_acc,'depth_acc':depth_acc}, "experiments/scribe_best.pt")
        print(f"  >>> Best! ({te_acc:.1f}%)")

    history.append({'epoch':epoch+1,'loss':tloss/nb,'train':tr_acc,'test':te_acc,
                    **{f'task_{t}':a for t,a in task_acc.items()},
                    **{f'depth_{d}':a for d,a in depth_acc.items()}})

print("\n"+"="*80)
print(f"Done! Best: {best_acc:.1f}%")

# ============================================================================
# PLOTS
# ============================================================================
import matplotlib.pyplot as plt
h = pd.DataFrame(history)
h.to_csv("experiments/scribe_history.csv", index=False)

fig,ax = plt.subplots(1,3,figsize=(18,5))
ax[0].plot(h['epoch'],h['loss'],'b-o',ms=4); ax[0].set_title('Loss'); ax[0].grid(alpha=0.3)
ax[1].plot(h['epoch'],h['train'],'b-o',ms=4,label='Train'); ax[1].plot(h['epoch'],h['test'],'r-s',ms=4,label='Test')
ax[1].set_title('Accuracy'); ax[1].legend(); ax[1].grid(alpha=0.3)
for c in [x for x in h.columns if x.startswith('task_')]:
    ax[2].plot(h['epoch'],h[c],'-o',ms=4,label=c.replace('task_',''))
ax[2].set_title('Per-Task'); ax[2].legend(); ax[2].grid(alpha=0.3)
plt.tight_layout(); plt.savefig("experiments/curves.png",dpi=150); plt.show()

b = torch.load("experiments/scribe_best.pt")
print(f"\nBest (Epoch {b['epoch']}): {b['test_acc']:.1f}%")
for t,a in sorted(b['task_acc'].items()): print(f"  {t}: {a:.1f}%")
for d,a in sorted(b['depth_acc'].items()): print(f"  {d}: {a:.1f}%")
```
